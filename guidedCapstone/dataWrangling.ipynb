```python
#Import pandas, matplotlib.pyplot, and seaborn in the correct lines below
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

```

```python
# load csv data 

credit_record = pd.read_csv("credit_record.csv")

credit_record
```

```python
#renamed columns

credit_record.rename(columns = {'MONTHS_BALANCE':'Balance','STATUS':'Status'}, inplace=True)

credit_record
```

```python
# groupby ID & Balance

credit_record.groupby(['ID'])['Balance'].count()

credit_record
```

```python
# upload application data

application_record = pd.read_csv("application_record.csv")
```

```python
application_record
```

```python
# remove Flag_Phone

application_record.drop(columns =['FLAG_PHONE'], inplace=True)

application_record
```

```python
# remove column

application_record.drop(columns =['FLAG_MOBIL'], inplace=True)

application_record
```

```python
# remove column

application_record.drop(columns =['FLAG_EMAIL'], inplace=True)

application_record
```

```python
# change nulls

application_record["OCCUPATION_TYPE"].fillna("Not Listed", inplace=True)

application_record
```

```python
# merge credit & app tables

cust_app = application_record.merge(credit_record, how='left', on='ID')
```

```python
 cust_app
```

```python
# check for nulls

cust_app.isnull().sum()
```

```python
cust_app["Balance"].fillna("Unavailable", inplace=True)
cust_app["Status"].fillna("Unavailable", inplace=True)
```

```python
cust_app.isnull().sum()
```

```python
cust_app
```

```python
# change column names

#renamed columns

cust_app.rename(columns = {'CODE_GENDER':'GENDER','FLAG_OWN_CAR':'Car','FLAG_OWN_REALTY':'Property','CNT_CHILDREN':'Children', 
                                     'AMT_INCOME_TOTAL':'Income', 
                                     'NAME_INCOME_TYPE':'Job', 'NAME_EDUCATION_TYPE':'Education','NAME_FAMILY_STATUS':'Martial_Status', 
                                     'NAME_HOUSING_TYPE':'Housing', 'DAYS_BIRTH':'Age','DAYS_EMPLOYED': 'Employed','OCCUPATION_TYPE':'Occupation', 'CNT_FAM_MEMBERS':'Family_Size'}, inplace=True)



cust_app

```

```python
# remove column

cust_app.drop(columns =['FLAG_WORK_PHONE'], inplace=True)

cust_app
```
